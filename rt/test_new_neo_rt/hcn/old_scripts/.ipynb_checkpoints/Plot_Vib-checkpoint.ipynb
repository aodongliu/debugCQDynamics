```python
import h5py
import matplotlib.pyplot as plt
import numpy

from functional import *


DAMP = 0.0007



def read_dataset(fname, path):
    f = h5py.File(fname)
    return numpy.array(f[path])

# Raw data
time = read_dataset('HCN_ccpVDZ_prot_8s8p8d_dx_0p001.bin', "RT/TIME")

ux = read_dataset('HCN_ccpVDZ_prot_8s8p8d_dx_0p001.bin', "RT/LEN_ELEC_DIPOLE")[:,0]
uy = read_dataset('HCN_ccpVDZ_prot_8s8p8d_dy_0p001.bin', "RT/LEN_ELEC_DIPOLE")[:,1]
uz = read_dataset('HCN_ccpVDZ_prot_8s8p8d_dz_0p001.bin', "RT/LEN_ELEC_DIPOLE")[:,2]

dt = time[1] - time[0]

signals = [ux, uy, uz]

# for s in signals:
#     s -= s[0]
#     s *= np.exp(-DAMP*time)


# Fourier transformed spectra
spects = []
freq = None
for s in signals:
    s -= s[0]
    #s *= np.exp(-DAMP*time)
    #w, f = fourier_tx(s, dt, 0)
    w, f = pade_tx(s, dt, wlim=(0,1), res=10000)
    freq = w
    spects.append(f)

total = sum([f.imag for f in spects]) / 3
total *= freq

# Plotting
plt.plot(freq*27.2114*8065.54429, total, c='k')
plt.xlim(2000,6500)
plt.ylim(0,0.002)
plt.ylabel("Intensity", fontsize=12)
plt.xlabel("Energy (cm$^{-1}$)", fontsize=12)
plt.title("HCN Proton vibrational spectra, RT-NEO-TDDFT/ccpVDZ, 8s8p8z")
plt.savefig("HCN_vib_spec.pdf")
plt.show()
```

```python
time = read_dataset('HCN_ccpVDZ_prot_8s8p8d_dx_0p001.bin', "RT/TIME")

ux = read_dataset('HCN_ccpVDZ_prot_8s8p8d_dx_0p001.bin', "RT/LEN_ELEC_DIPOLE")[:,0]

# Input window cutoff in a.u.
cutoff = 100


window_size = int(cutoff/dt)

window_signal_size = len(time) - window_size

windowed_signal = [np.average(ux[i:(window_size+i)]) for i in range(window_signal_size)]
windowed_time = time[window_size:]

# Plot
f, ax = plt.subplots(nrows=2, sharex=False)
ax[0].plot(time, ux)  # Raw signal on top subplot
ax[0].set_xlabel("Time", fontsize=14)
ax[0].set_ylabel("Dipole (a.u.)", fontsize=14)


ax[1].plot(windowed_time, windowed_signal)  # Windowed average on bottom subplot
# Compare windowed average to the "correct" component of the signal
# Phase differences aren't super important - it's just to look at the signal
ax[1].set_xlabel("Time (a.u)", fontsize=14)

ax[1].set_ylabel("Dipole (a.u.)", fontsize=14)
plt.tight_layout()
plt.savefig("X_dipole_window_average.pdf")
plt.savefig("window_avg_example.pdf")
plt.show()
```
